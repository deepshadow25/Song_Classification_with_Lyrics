```python
import pandas as pd
```

```python
df_melon = pd.read_csv('melon_lyrics.csv', encoding = 'cp949')
df_bugs = pd.read_csv('bugs_lyrics.csv')
df_genie = pd.read_csv('bugs_lyrics.csv')
```

```python
df_bugs.columns
```

```python
del df_genie['Unnamed: 0']
del df_bugs['Unnamed: 0']
del df_bugs['url']
del df_genie['url']
df_bugs = df_bugs[['year','singer','names','lyrics']]
df_genie.columns = [['names','singer','year','lyrics']]
df_genie = df_genie[['year','singer','names','lyrics']]
```

```python
def remove_space(df):
  for i in range(len(df)):
    df['names'][i] = df['names'][i].strip().replace(' ','')
    df['singer'][i] = df['singer'][i].strip()
    df['lyrics'][i] = df['lyrics'][i].replace('\n', '')
```

```python
def find_rep(df1, df2):
  #song_list = [song2 for song2 in df2['names'] if song2 not in list(df1['names']))]
  

  ##혹은 밑에처럼. 개념은 같다
    song_list = []
    for index in range(len(df2)):
        print(df2.loc[index])
        if df2.loc[index]['names'] not in list(df1['names']) and df2.loc[index]['singer'] not in list(df1['singer']):
                song_list.append(df2.loc[index])
    return song_list


#find_rep(df_melon, df_bugs)
```

```python
remove_space(df_melon)
remove_space(df_bugs)
remove_space(df_genie)
```

```python
song1 = find_rep(df_melon, df_bugs) #멜론과 벅스에서 안 겹치는 벅스노래
song2 = find_rep(df_melon, df_genie) #멜론과 지니에서 안 겹치는 벅스노래
song_final = list(df_melon['names']) +song1 + song2 #멜론과 안 겹치는 노래들을 합친 리스트
print(song_final)
len(song_final)
```
